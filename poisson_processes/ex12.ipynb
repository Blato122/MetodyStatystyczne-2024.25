### 12. Symulacja procesu Poissona

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def generate_time(lambda1):
    u = random.random()
    return -np.log(u)/lambda1
```

```python
def simulate_poisson(lambda1, tmin, tmax):
    events = []
    current_time = tmin
    while current_time < tmax:
        events.append(current_time)
        current_time += generate_time(lambda1)
    return events
```

```python
lambda1 = 1 # one event per minute
tmin, tmax = 0, 90
n_trajectories = 12
```

```python
trajectories = \
[simulate_poisson(lambda1, tmin, tmax) for _ in range(n_trajectories)]
```

```python
plt.figure(figsize=(10, 6))
for traj in trajectories:
    time = list(range(len(traj)))
    plt.step(traj, time, where='post')

plt.title("Zasymulowanie trajektorie procesu Poissona")
plt.xlabel("Czas (minuty)")
plt.ylabel("Zliczenia")
plt.grid()
plt.show()
```

```python
for traj in trajectories:
    print(traj[1], traj[20], max(traj))
```

Proces Poissona jest stacjonarny od samego pocztku. Liczby zdarze w dwch rozcznych przedziaach czasowych s niezalenymi zmiennymi losowymi. Proces Poissona nie ma pamici, co oznacza, e wczeniejsze zdarzenia nie wpywaj na prawdopodobiestwo zajcia zdarzenia w danym czasie.
